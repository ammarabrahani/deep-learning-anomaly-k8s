```python
# Step 1: Preprocessing
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Load and scale all features
df = pd.read_csv('../data/raw/creditcard.csv')
df = df.drop(columns=['Time'])

X = df.drop(columns=['Class'])
y = df['Class']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, stratify=y, random_state=42
)

# Prepare Autoencoder data
import numpy as np
X_train_auto = X_train[y_train == 0]
X_test_auto = X_test
y_test_auto = y_test

print(f'Train shape: {X_train.shape}, Test shape: {X_test.shape}')
print('Train label counts:', y_train.value_counts())
print('Test label counts:', y_test.value_counts())
```

```python
# Step 2: Train Deeper Autoencoder
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras import regularizers

input_dim = X_train_auto.shape[1]

# Deeper architecture
input_layer = Input(shape=(input_dim,))
encoded = Dense(32, activation='relu')(input_layer)
encoded = Dense(16, activation='relu')(encoded)
bottleneck = Dense(8, activation='relu')(encoded)
decoded = Dense(16, activation='relu')(bottleneck)
decoded = Dense(32, activation='relu')(decoded)
output_layer = Dense(input_dim, activation='linear')(decoded)

autoencoder = Model(inputs=input_layer, outputs=output_layer)
autoencoder.compile(optimizer='adam', loss='mse')

# Train the model
autoencoder.fit(X_train_auto, X_train_auto,
                epochs=50,
                batch_size=128,
                shuffle=True,
                validation_split=0.1)
```

```python
# Step 3: Evaluate Model
from sklearn.metrics import classification_report, f1_score
import matplotlib.pyplot as plt

reconstructions = autoencoder.predict(X_test_auto)
mse = np.mean(np.square(X_test_auto - reconstructions), axis=1)

# Visualize MSE
plt.figure(figsize=(10, 4))
plt.hist(mse, bins=100)
plt.title('Reconstruction Error Distribution')
plt.xlabel('MSE')
plt.ylabel('Frequency')
plt.show()

# Use lower threshold (90th percentile)
threshold = np.percentile(mse, 90)
print(f'Threshold (90th percentile): {threshold:.5f}')

y_pred = (mse > threshold).astype(int)

# Classification Report
print('Classification Report (Autoencoder):')
print(classification_report(y_test_auto, y_pred, target_names=['Normal', 'Anomaly']))

f1_macro = f1_score(y_test_auto, y_pred, average='macro')
print(f'Macro Avg F1-score: {f1_macro:.4f}')
```
