Description:
This script compares the performance of multiple anomaly detection algorithms, including:
- Deep Q-Learning (via a custom wrapper class: `RealDeepQLWrapper`)
- Isolation Forest
- Local Outlier Factor (LOF)
- One-Class SVM (OCSVM)
- SGD-based One-Class SVM


Each model is applied to detect anomalies in datasets with a known injected fraction of outliers (15%). The models are evaluated using the **Macro F1-score**, which is printed for each algorithm per dataset. The decision boundaries and point classifications are plotted using Matplotlib.

Key functionality includes:
- Meshgrid creation for decision boundary plotting
- Timing of model fitting
- Visual comparison across multiple datasets and algorithms
- Saving the final comparative plot

This visualization helps understand the **adaptability and robustness** of each model to different data geometries.

```python
import os
import time
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs, make_moons
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
from sklearn.svm import OneClassSVM
from sklearn.linear_model import SGDOneClassSVM

# If the real_dql_model is in another folder, adjust the path accordingly
import sys
sys.path.append("../src")  # Update this if your file is elsewhere
from real_dql_model import RealDeepQLWrapper
from sklearn.metrics import f1_score
```

```python
# Configuration
n_samples = 300
outliers_fraction = 0.15
n_outliers = int(outliers_fraction * n_samples)
n_inliers = n_samples - n_outliers

anomaly_algorithms = [
    ("Deep Q-Learning", RealDeepQLWrapper(episodes=20, arch='wide')),
    ("One-Class SVM", OneClassSVM(nu=0.15, kernel="rbf", gamma=0.1)),
    ("One-Class SVM (SGD)", SGDOneClassSVM(nu=0.15, random_state=42)),
    ("Isolation Forest", IsolationForest(contamination=outliers_fraction, random_state=42)),
    ("Local Outlier Factor", LocalOutlierFactor(n_neighbors=35, contamination=outliers_fraction)),
]

blobs_params = dict(random_state=0, n_samples=n_inliers, n_features=2)
datasets = [
    make_blobs(centers=[[0, 0], [0, 0]], cluster_std=0.5, **blobs_params)[0],
    make_blobs(centers=[[2, 2], [-2, -2]], cluster_std=[0.5, 0.5], **blobs_params)[0],
    make_blobs(centers=[[2, 2], [-2, -2]], cluster_std=[1.5, 0.3], **blobs_params)[0],
    4.0 * (make_moons(n_samples=n_samples, noise=0.05, random_state=0)[0] - np.array([0.5, 0.25])),
    14.0 * (np.random.RandomState(42).rand(n_samples, 2) - 0.5),
]
```

```python
xx, yy = np.meshgrid(np.linspace(-7, 7, 150), np.linspace(-7, 7, 150))

plt.close('all')
fig = plt.figure(figsize=(len(anomaly_algorithms) * 2 + 4, 12.5))
plt.subplots_adjust(left=0.02, right=0.98, bottom=0.001, top=0.96, wspace=0.05, hspace=0.01)

plot_num = 1
rng = np.random.RandomState(42)

for i_dataset, X in enumerate(datasets):
    X = np.concatenate([X, rng.uniform(low=-6, high=6, size=(n_outliers, 2))], axis=0)

    for name, algorithm in anomaly_algorithms:
        if name == "Deep Q-Learning" and i_dataset > 0:
            continue

        t0 = time.time()
        algorithm.fit(X)
        t1 = time.time()

        ax = fig.add_subplot(len(datasets), len(anomaly_algorithms), plot_num)
        if i_dataset == 0:
            ax.set_title(name, size=14)

        if name == "Local Outlier Factor":
            y_pred = algorithm.fit_predict(X)
        else:
            y_pred = algorithm.predict(X)

        
        # Simulate ground truth
        y_true = np.ones_like(y_pred)
        y_true[-n_outliers:] = -1

        f1 = f1_score(y_true, y_pred, average='macro')
        print(f"ðŸ“Š Macro F1-score ({name}) on dataset {i_dataset + 1}: {f1:.4f}")

        if name != "Local Outlier Factor":
            Z = algorithm.predict(np.c_[xx.ravel(), yy.ravel()])
            Z = Z.reshape(xx.shape)
            ax.contour(xx, yy, Z, levels=[0], linewidths=2, colors="black")

        colors = np.array(["#377eb8", "#ff7f00"])
        ax.scatter(X[:, 0], X[:, 1], s=10, color=colors[(y_pred + 1) // 2])
        ax.set_xlim(-7, 7)
        ax.set_ylim(-7, 7)
        ax.set_xticks(())
        ax.set_yticks(())
        ax.text(0.99, 0.01, ("%.2fs" % (t1 - t0)).lstrip("0"),
                transform=ax.transAxes, size=12, ha="right")

        plot_num += 1

# Save and show
save_path = "comparison_with_real_dql.png"
fig.savefig(save_path, dpi=300)
plt.show()
print(f"âœ… Plot saved as '{save_path}'")
```
